Q3. What is App routing in Flask? Why do we use app routes?

Ans:---

1)In Flask, app routing refers to the process of defining URL routes for different parts of your web application. It allows us to map specific URLs (or routes) to specific Python functions, known as view functions, that handle the logic for generating the content to be displayed when a user accesses those URLs.

2)In Flask,we define app routes using the **@app.route()** decorator, which binds a URL pattern to a Python function.

3)We use app routes in Flask to map specific URLs to Python functions (view functions) that handle the logic for generating the content to be displayed when users access those URLs. App routes enable URL mapping, modularization of code, dynamic URLs, and handling different HTTP methods, making it easy to manage user requests and generate appropriate responses in web applications.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans:---

In Flask, the **url_for()** function is used for URL building.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return "Hello, this is the homepage!"

@app.route('/user/<username>')
def show_user_profile(username):
    return f"Hello, {username}!"

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f"This is post number {post_id}."

if __name__ == '__main__':
    with app.test_request_context():
    
        homepage_url = url_for('index')
        user_url = url_for('show_user_profile', username='JohnDoe')
        post_url = url_for('show_post', post_id=123)


    print("Homepage URL:", homepage_url)
    print("User URL:", user_url)
    print("Post URL:", post_url)
```
